1) Crie uma função que gere n partículas aleatórias distribuidas uniformemente dentro de uma área minx, miny, maxx, maxy (passados como parâmetros). Veja a funcao
Devolve uma lista com n partículas

```python
#var_theta = math.pi/3
def nb_initialize_particle_cloud(pose, minx, maxx, miny, maxy, var_theta, n):
    particle_cloud = []
    s = pose
    var_x = maxx - minx
    var_y = maxy-miny
    for i in range(n):
        x = random.uniform(s[0] - var_x, s[0] + var_x)
        y = random.uniform(s[1] - var_x, s[1] + var_y)
        theta = random.uniform(s[2] - var_theta, s[2] + var_theta)
        p = Particle(x, y, theta, w=1.0) # A prob. w vai ser normalizada depois
        particle_cloud.append(p)
    return particle_cloud
```

5) Crie uma função que aplica um deslocamento [delta_x, delta_y, delta_theta] com um desvio padrão [std_x, std_y,
std_theta] a todas as partículas

```python
def deslocamento (movimentos, delta, std):
    movs=[]
    for mov in movimentos:
        movN = []
        movN[0] += delta[0] + np.random.randint(0, std[0])
        movN[1] += delta[1] + np.random.randint(0, std[1])
        movN[2] += delta[2] + np.random.randint(0, std[2])
        movs.append(movN)
    return movs
```
